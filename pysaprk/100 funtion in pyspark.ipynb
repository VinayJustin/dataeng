```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
spark = SparkSession.builder.appName("SparkOperations").getOrCreate()
df = spark.read.options(header='True', inferSchema='True').csv('/Volumes/pyspark/emp/pyspark/employee_data.csv')
df.display()


```

```python
df.select("name").show(3)
```

```python
df.filter(df.age > 40 ).show()
```

```python
df.withColumn("total_inHand", df.salary+ df.bonus).show(truncate=False)
```

```python
df.withColumnRenamed("id","employee_id").show()
```

```python
df.orderBy(df.age.desc()).show()
```

```python
df.count()
```

```python

df.withColumn("constant_col",lit("constant")).show()
```

```python
##cast()

df.withColumn("bonus",df.bonus.cast("string")).printSchema()
```

```python
df.withColumn("bonus",df.bonus.cast("int") + 100 ).show()
```

```python
## with().otherwise()

from pyspark.sql.functions import when
df.withColumn("high_bonus", when(df.bonus > 50000, True ).otherwise(False)).show()
```

```python
df.show()
```

```python
df.select(df.bonus.alias("performance_bonus")).show()
```

```python
df.withColumn("department_age",concat(df.department,lit("_"),df.age)).show()
```

```python
df.withColumn("Name", upper(df.name)).show()
```

```python
df.withColumn("is_null",df.skills.isNull()).show()
```
